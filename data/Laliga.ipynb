```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#reading csv file
laliga=pd.read_csv("2018laliga.csv")
laliga.head(20)
```

```python
#exploring columns
laliga.shape
```

```python
#exploring null data
laliga.info()
```

```python
#Teams with games done and won
laliga[['Team', 'matches', 'wins', 'loses']].plot(kind='bar', title='Games won by teams')
```

```python
#Teams analyzing shots vs on_target
laliga[['Team','Shots', 'S_OnTarget']].plot(kind='bar', title='Shots vs on_target')
```

```python
#Teams with more than 15 games won
morethan15gameswins = laliga.where(laliga['wins']>=15)
morethan15gameswins[['Team', 'matches', 'wins']].plot.bar(title='Teams with more than 15 games won')
```

```python
#Top 5 Teams with more than 15 games won
morethan15gameswins = laliga.where(laliga['wins']>=15)
morethan15gameswins[['Team', 'matches', 'wins']].head(6)
```
