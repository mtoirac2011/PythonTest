```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
%ls
```

```python
#reading csv file
rm_players=pd.read_csv("real_madrid_fc.csv")
rm_players.head(20)
```

```python
#exploring columns
rm_players.shape
```

```python
#exploring null data
rm_players.info()
```

```python
# converting values of 'Player' 'Nation' 'Pos' columns into string datatype
rm_players['Player'] = rm_players['Player'].astype('str')
rm_players['Nation'] = rm_players['Nation'].astype('str')
rm_players['Pos'] = rm_players['Pos'].astype('str')

#I am not able to see the new data type columns converted 
```

```python
#exploring null data after converting data types
rm_players.info()
```

```python
#analizing how data are distributed
rm_players.describe()
```

```python
#grouping player by nations
rm_players.groupby(['Nation']).count()['Id']
```

```python
#grouping player by nations (Graphical)
rm_players.groupby(['Nation']).count()['Id'].plot.bar(title='Player by nations')
```

```python
#calculating median of player's Age
rm_players['Age'].median()
```

```python
#histogram by Age
rm_players.Age.plot(kind='hist', title='Histogram for Age', color='c')
```

```python
#calculating players with more than 27
oldplayers = rm_players[rm_players['Age'].between(28,40)]
oldplayers[['Player','Age']]
```

```python
#sort descending
rm_players.sort_values(by='Age',inplace=True,ascending=False)
rm_players[['Player', 'Age']]
```
